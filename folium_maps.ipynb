```python
import pandas as pd

# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lon':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5],
   'lat':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97],
   'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador'],
   'value':[10, 12, 40, 70, 23, 43, 100, 43]
}, dtype=str)

data
```

```python
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)

# add marker one by one on the map
for i in range(0,len(data)):
    html=f"""
        <h1> {data.iloc[i]['name']}</h1>
        <p>You can use any html here! Let's do a list:</p>
        <ul>
            <li>Item 1</li>
            <li>Item 2</li>
        </ul>
        </p>
        <p>And that's a <a href="https://python-graph-gallery.com">link</a></p>
        """
    iframe = folium.IFrame(html=html, width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
        popup=popup,
    ).add_to(m)
m
```

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
select_widget = widgets.Select(
    options=['Open Street Map', 'Terrain', 'Toner', 'Watercolor', 'Positron', 'Dark Matter'],
    value='Open Street Map',
    description='Map Type:',
    disabled=False
)

# Widget function
def select(map_type):
    if map_type == 'Open Street Map':
        m = folium.Map(location=[20, 0], tiles="OpenStreetMap", zoom_start=2)
    elif map_type == 'Terrain':
        m = folium.Map(location=[20, 0], tiles='Stamen Terrain', zoom_start=2)
    elif map_type == 'Toner':
        m = folium.Map(location=[20, 0], tiles='Stamen Toner', zoom_start=2)
    elif map_type == 'Watercolor':
        m = folium.Map(location=[20, 0], tiles='Stamen Watercolor', zoom_start=2)
    elif map_type == 'Positron':
        m = folium.Map(location=[20, 0], tiles='CartoDB Positron', zoom_start=2)
    elif map_type == 'Dark Matter':
        m = folium.Map(location=[20, 0], tiles='CartoDB Dark_Matter', zoom_start=2)
    
    # Adding markers
    for i in range(len(data)):
        html = f"""
            <h1>{data.iloc[i]['name']}</h1>
            <p>You can use any html here! Let's do a list:</p>
            <ul>
                <li>Item 1</li>
                <li>Item 2</li>
            </ul>
            <p>And that's a <a href="https://python-graph-gallery.com">link</a></p>
        """
        iframe = folium.IFrame(html=html, width=200, height=200)
        popup = folium.Popup(iframe, max_width=2650)
        folium.Marker(
            location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
            popup=popup,
        ).add_to(m)
    
    display(m)

# Interaction between widget and function
widgets.interact(select, map_type=select_widget)
```

```python
airports=pd.read_csv('ca-airports.csv')
airports
```

```python
airports.drop(1)
airports.head(6)
```

```python


# Display the modified DataFrame
ports=airports.head(6)
```

```python
# multiple markers using dataframe
# there is an example below using apply function instead of loop

# create map
map_ports = folium.Map(location=[20, 0], zoom_start=4)

# plot airport locations
for (index, row) in ports.iterrows():
    folium.Marker(location=[row.loc['latitude_deg'], row.loc['longitude_deg']], 
                  popup=row.loc['name'], 
                  tooltip='click').add_to(map_ports)
    
# display map    
map_ports
```

```python
# search for 'glyphicon bootstrap' icons to see examples



map_ports = folium.Map(location=[ports.iloc[0]['latitude_deg'], ports.iloc[0]['longitude_deg']], zoom_start=6)

# Iterate over the DataFrame rows and add markers
for index, row in ports.iterrows():
    folium.Marker(
        location=[row['latitude_deg'], row['longitude_deg']],
        popup=row['name'],
        tooltip='click',
        icon=folium.Icon(icon='plane', prefix='glyphicon')
    ).add_to(map_ports)

# Display the map
map_ports
```
